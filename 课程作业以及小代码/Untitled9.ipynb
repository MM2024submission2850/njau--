```python
import re
from pyquery import PyQuery as pq
from collections import defaultdict

def readtxt():
    doc=[]
    entity_dict=defaultdict(list)
    with open(r'O:\大三下\SRT\animal_food_safety_data.txt','r',encoding='utf-8') as f:
        for line  in f.readlines():
            if line.strip()!='':
                doc.append(line.strip())
    for i in range(len(doc)):
        new_line = re.sub("(?<!<)/[A-Za-z]+[0-9]*", "", doc[i])      #(?<!exp2)exp1 查找前面不是exp2的exp1
        final=re.sub(" ","",new_line)
        for o in pq("<doc>{}</doc>".format(final)).contents():
            if hasattr(o, "tag"):
                text = pq(o).text()
                entity_dict[o.tag].append(text)
    print(entity_dict)
    keys=entity_dict.keys()
    print(list(keys))
    for key in keys:
        values=list(entity_dict[key])
        with open('O:\\大三下\\SRT\\'+key+'.txt','w',encoding='utf-8') as f:
            for value in values:
                f.write(value+'\n')

def main():
    readtxt()

if __name__ == '__main__':
    main()
```

```python

import os, codecs
import jieba
from collections import Counter
 
def get_words(txt):
    seg_list = jieba.cut(txt)
    c = Counter()
    for x in seg_list:
        if len(x)>1 and x != '\r\n':
            c[x] += 1
    print('常用词频度统计结果')
    for (k,v) in c.most_common(500):
        print(k, v)
 
if __name__ == '__main__':
    with codecs.open(r'O:\大三下\SRT\animal_food_safety_data.txt', 'r', 'utf8') as f:
        txt = f.read()
    get_words(txt)
```
