```python
import plotly as py
import plotly.graph_objs as go
import numpy as np
import csv
from matplotlib import pyplot as plt
from datetime import datetime


pyplt = py.offline.plot

filename = 'Overall.csv'
with open(filename) as f:
    reader = csv.reader(f)
    heeader_row = next(reader)
    date = []
    currentConfirmedCount = []  # 现存确诊人数累计死亡人数
    confirmedCount = []   # 累计确诊人数（今日新增确诊）
    curedCount = []  # 累计治愈人数
    deadCount = []  # 累计死亡人数

    for row in reader:
        new_date = row[4]
        date.append(new_date)
        currentConfirmed = int(row[0])
        currentConfirmedCount.append(currentConfirmed)
        confirmed = int(row[1])
        confirmedCount.append(confirmed)
        cured = int(row[2])
        curedCount.append(cured)
        dead = int(row[3])
        deadCount.append(dead)

trace1 = go.Scatter(
    x = date,
    y = currentConfirmedCount,
    fill='tonexty',
    mode='none',  # 无边界线
    name="现存确诊人数累计死亡人数"
)
trace2 = go.Scatter(
    x = date,
    y = confirmedCount,
    fill='tozeroy',
    mode='none',  # 无边界线
    name="累计确诊人数"
)

trace3 = go.Scatter(
    x = date,
    y = curedCount,
    fill='tozeroy',
    mode='none',  # 无边界线
    name="累计治愈人数"
)

trace4 = go.Scatter(
    x = date,
    y = deadCount,
    fill='tozeroy',
    mode='none',  # 无边界线
    name="累计死亡人数"
)

data = [trace1, trace2, trace3, trace4]

layout = dict(title='中国受疫情影响人数走势',
              xaxis=dict(title='时期'),
              yaxis=dict(title='单位：人'),
              )
fig = dict(data=data, layout=layout)
pyplt(fig, filename='2.html')
```
