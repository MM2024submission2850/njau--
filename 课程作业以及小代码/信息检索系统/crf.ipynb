```python

import random
import os

def read(path):
    with open(path, 'r', encoding = 'utf-8', errors = 'ignore')as f:
        lines = f.read()
    return lines
#写入文本
def write(data,path):
    with open(path,'a',encoding='utf-8') as f:
        f.write(data)
#将训练语料和测试语料分出来
def corpus(word_list,path):
    for word in word_list:#word是词word_list是一句话
        if word==' ':
            pass
        elif word=='  ':
            pass
        elif word=='\n':
            pass
        elif len(word)==1:
            write(word+'\t'+'S-seg'+'\n', path)
        elif len(word)>=2:
            length=len(word)
            write(word[0] + '\t' + 'B-seg' + '\n', path)
            for j in range(1,length-1):
                write(word[j]+'\t'+'I-seg'+'\n', path)
            write(word[length-1] +'\t'+ 'E-seg' + '\n', path)
    write('\n',path)
def main():
    list=read(r'O:\大三下\信息检索CRF\jiaodui.txt')
    list = list.split('\n')
    random.shuffle(list)  # 将每个句子打乱
    length = len(list)  # length是句子的个数
    train_length = int(length / 10 * 9 - 1)  # 按照9:1分成训练预料和测试语料
    while ' ' in list:
        list.remove(' ')
    for item in list[0:train_length]:#按照格式写入训练语料
        item.strip(" ")  # 删除每句话开头和结尾的空格
        item=item.split("/")#item是句子,将每个句子按照'/'进行分割
        corpus(item , r'O:\大三下\信息检索CRF\train.txt')
    for item in list[train_length:length]:#按照格式写入测试预料
        item.strip(" ")  # 删除每句话开头和结尾的空格
        item = item.split('/')  # item是句子,将每个句子按照'/'进行分割
        corpus(item , r'O:\大三下\信息检索CRF\test.txt')


if __name__=="__main__":
  main()


```
