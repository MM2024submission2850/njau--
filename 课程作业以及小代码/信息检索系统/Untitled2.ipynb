```python
#定义文件系统类
class wx_manage:
    #初始化
    def __init__(self):
        #初始化存储数组
        self.data = []
        #读取txt文件
        with open(r'C:\Users\13284\Desktop\fenxi\data.txt','r',encoding='utf-8') as f:
            series = f.readline()
            while series and series!='':
                info = series.split('\t')
                doi = info[0]
                title = info[1]
                name = info[2]
                year = int(info[3])
                abstract = info[4][:-1]
                wx_info = [doi, title, name, year, abstract]
                self.data.append(wx_info)
                series = f.readline()
    #功能菜单
    def Query_menu(self):
        print('    1>add  2>del  3>search  4>modify   5>show 0>menu  -1>quit  ')
        while True:
            try:
                opition=int(input('请输入功能序号，以回车键结束：'))
                if opition==1:
                    self.add()
                    break
                elif opition==2:
                    self.Del()
                    break
                elif opition==3:
                    self.search()
                    break
                elif opition==4:
                    self.modify()
                    break
                elif opition==5:
                    self.show()
                    break
                elif opition==0:
                    self.Query_menu()
                    break
                elif opition==-1:
                    return -1
                else:
                    print('请输入正确的功能序号！')
                    continue
            except:
                print('请输入正确的功能序号！')
                continue
        return 0
    #添加文献信息
    def add(self):
        while True:
            info=input('请输入要添加的文献信息,格式：doi 标题 作者 出版年份 摘要（以空格间隔，回车结束）：').split(' ')
            #格式检验
            if len(info) != 5:
                print('请按指定格式输入！')
                continue
            else:
                #出版年份格式检验
                try:
                    info[3]=int(info[3])
                    break
                except:
                    print('请按指定格式输入！')
                    continue
        flag=self.search(doi=info[0],add=True)
        if flag==-1:
            self.data.append([info[0],info[1],info[2],info[3],info[4]])
            self.saveback()
            print('添加成功！')
            print('已有数据：%d'%len(self.data))
        else:
            print('该文献信息已存在！')
    #删除文献信息
    def Del(self):
        doi=input('请输入要删除文献的doi：')
        #查询信息
        flag=self.search(doi=doi)
        if flag!=-1:
            del self.data[flag]
            self.saveback()
            print('删除成功！')
    #查找文献信息，ID为面对Del函数的可选变量,add为面向add函数的可选变量
    def search(self,doi='search',add=False):
        if doi=='search':
            doi=input('请输入要查找文献的doi：')
        flag=-1
        n=-0
        for info in self.data:
            if info[0]==doi and add==False:
                print('doi\t期刊\t作者\t出版年份\t摘要')
                print(info[0]+'\t'+info[1]+'\t'+info[2]+'\t'+str(info[3])+'\t'+info[4])
                flag=n
            elif info[0]==doi and add==True:
            	flag=n
            n+=1
        if flag==-1 and add==False:
            print('文献信息不存在！')
            return -1
        else:
            return flag
    #修改文献信息（出版年份+1）
    def modify(self):
        doi=input('请输入要修改文献的doi：')
        #查询信息
        flag=self.search(doi=doi)
        if flag!=-1:
            del self.data[flag]
            info=input('请输入要添加的文献信息,格式：doi 标题 作者 出版年份 摘要（以空格间隔，回车结束）：').split(' ')
            self.data.append([info[0],info[1],info[2],info[3],info[4]])
            self.saveback()
        print('修改成功')
    #展示文献信息
    def show(self):
        print('doi\t期刊\t作者\t出版年份\t摘要')
        for info in self.data:
            print(info[0] + '\t' + info[1] + '\t' + info[2] + '\t' + str(info[3]) + '\t' + info[4])
    #保存
    def saveback(self):
        with open(r'C:\Users\13284\Desktop\fenxi\data.txt','w+',encoding='utf-8') as f:
            for info in self.data:
                f.write(info[0]+'\t'+info[1]+'\t'+info[2]+'\t'+str(info[3])+'\t'+info[4]+'\n')


if __name__ == '__main__':
    Sys=wx_manage()
    #建立退出标志
    flag=0
    #开始循环运行
    while flag==0:
        flag=Sys.Query_menu()
        if flag==0:
            check=input('继续操作请回车：')
```
