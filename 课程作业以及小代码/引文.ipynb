```python
import re
import pandas as pd
import numpy as np
import xlwt

df = pd.read_excel(r"O:\大三下暑假\结构拆分数据scientometrics\数据处理\scientometrics_final_labeled.xlsx")
aa = []  #创建一个空列表
bb=[]
print(len(df['paragraph']))
for j in range(len(df['paragraph'])):
    #bb = re.findall("\【#bib\d+\s\w\w\w\d+\s\w\w\w+\s",str(df['paragraph'][j]))
    bb = re.findall("\【#CR[1-9][0-9]?\】",str(df['paragraph'][j]))
    #bb = re.findall("\【#bib\d+",str(df['paragraph'][j]))
    aa.append(bb)
    print (aa[j])  #查看 bb[0]

f = xlwt.Workbook()               #创建工作簿
sheet1 = f.add_sheet(u'sheet1',cell_overwrite_ok=True) #创建sheet
for i in range(0,len(aa)):
    sheet1.write(i,0,aa[i])    #write的第一个,第二个参数代表单元格的坐标, 第三个是要写入的数
f.save(r"C:\Users\13284\Desktop\2.xls")#保存文件
print ('已处理完毕！！！！')
```
