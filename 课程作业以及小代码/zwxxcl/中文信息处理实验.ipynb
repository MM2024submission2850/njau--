```python
import os
import pandas as pd
dir_path = r"O:\大三下\中文信息处理实验\陈昱成\\"
save_path = r"O:\大三下\中文信息处理实验\\"

all_excel = os.listdir(r"O:\大三下\中文信息处理实验\陈昱成\\")

for each in all_excel:
    file_path = os.path.join(dir_path, each)
#     print(file_path)
    file = pd.read_excel(file_path)
    
    full_text = ""
    for i in range(len(file)):
        tag = file.at[i,"分类标记"]
        text = str(file.at[i,"原文"])
        if tag == "e":
            text += "\n"
        full_text += text
        
#     print(full_text)
    #保存到txt
    with open(os.path.join(save_path, each.replace("xls","txt")), "wt", encoding="utf8") as txt:
        txt.write(full_text)
```

```python
biaotou=['序号','原文','长度','该行与下一行的长度差值','分类标记']
#在哪里搜索多个表格
filelocation=r"O:\大三下\中文信息处理实验\陈昱成\\"
#当前文件夹下搜索的文件名后缀
fileform="xls"
#将合并后的表格存放到的位置
filedestination=r"C:\Users\13284\Desktop\\"
#合并后的表格命名为file
file="test"
 
#首先查找默认文件夹下有多少文档需要整合
import glob
from numpy import *
filearray=[]
for filename in glob.glob(filelocation+"*."+fileform):
    filearray.append(filename)
#以上是从pythonscripts文件夹下读取所有excel表格，并将所有的名字存储到列表filearray
print("在默认文件夹下有%d个文档哦"%len(filearray))
ge=len(filearray)
matrix = [None]*ge
#实现读写数据
 
#下面是将所有文件读数据到三维列表cell[][][]中（不包含表头）
import xlrd
for i in range(ge):
    fname=filearray[i]
    bk=xlrd.open_workbook(fname)
    try:
        sh=bk.sheet_by_name("sheet1")
    except:
        print ("在文件%s中没有找到sheet1，读取文件数据失败,要不你换换表格的名字？" %fname)
    nrows=sh.nrows 
    matrix[i] = [0]*(nrows-1)
    
    ncols=sh.ncols
    for m in range(nrows-1):  
        matrix[i][m] = ["0"]*ncols
 
    for j in range(1,nrows):
        for k in range(0,ncols):
            matrix[i][j-1][k]=sh.cell(j,k).value
#下面是写数据到新的表格test.xls中哦
import xlwt
filename=xlwt.Workbook()
sheet=filename.add_sheet("原文")
#下面是把表头写上
for i in range(0,len(biaotou)):
    sheet.write(0,i,biaotou[i])
#求和前面的文件一共写了多少行
zh=1
for i in range(ge):
    for j in range(len(matrix[i])):
        for k in range(len(matrix[i][j])):
            sheet.write(zh,k,matrix[i][j][k])
        zh=zh+1
print("我已经将%d个文件合并成1个文件，并命名为%s.xls.快打开看看正确不？"%(ge,file))
filename.save(filedestination+file+".xls")
```

```python
# 实体标注
def entity(path):
    filepath = "entity.txt"
    baseio.writetxt_w("", filepath)
    text = baseio.read_txt(path)
    # print(text)
    text = text.replace(" ", "")
    text = text.replace("/。/\n", "/。/\n/")
    textlist = text.split("/")
    print(textlist)
    for word in textlist:  # 进行标注处理
        if word == '\n':
            baseio.writetxt_a(str(word), filepath)
        elif len(word) == 1 and word != ('\n' or '' or ' '):
            txt_data = "{0}{1}{2}{3}".format(word, '\t', "S-seg", '\r\n')
            # print(txt_data)
            baseio.writetxt_a(str(txt_data), filepath)
        elif len(word) == 2:
            txt_data1 = "{0}{1}{2}{3}".format(word[0], '\t', "B-seg", '\r\n')
            # print(txt_data1)
            baseio.writetxt_a(str(txt_data1), filepath)
            txt_data2 = "{0}{1}{2}{3}".format(word[1], '\t', "E-seg", '\r\n')
            # print(txt_data2)
            baseio.writetxt_a(str(txt_data2), filepath)
        elif len(word) > 2:
            txt_data1 = "{0}{1}{2}{3}".format(word[0], '\t', "B-seg", '\r\n')
            # print(txt_data1)
            baseio.writetxt_a(str(txt_data1), filepath)
            for i in range(1, len(word) - 1):
                txt_data2 = "{0}{1}{2}{3}".format(word[i], '\t', "I-seg", '\r\n')
                # print(txt_data2)
                baseio.writetxt_a(str(txt_data2), filepath)
            txt_data3 = "{0}{1}{2}{3}".format(word[len(word) - 1], '\t', "E-seg", '\r\n')
            # print(txt_data3)
            baseio.writetxt_a(str(txt_data3), filepath)
```

```python
import os
import codecs

def fun(path):
    fileArray = []    
    for root, dirs, files in os.walk(path):
        for fn in files:
            eachpath = str(root+'/'+fn)
            fileArray.append(eachpath)
    return fileArray 

def readfile(raw_file_path):
    with open(raw_file_path,"rt",encoding="utf8",errors="ignore") as f:
        raw_file=f.readlines()
        return raw_file
    
def writetoken2txt(questionlist, path):
    tokenlist = []
    for each in questionlist:
        # print(each)
        eachlist = list(each)
        print(eachlist)
        count = 0
        for i in range(len(eachlist)):
            i = count
            if i < len(eachlist):
                if eachlist[i] == "【":
                    eachlist[i+1] = eachlist[i+1] + "\t" + "B-shiti1" + "\n"
                    tokenlist.append(eachlist[i+1])
                    for j in range(len(eachlist[i+2:])):
                        if j < len(eachlist[i+2:]):
                            if eachlist[i+2+j] != "】":
                                eachlist[i+2+j] = eachlist[i+2+j] + "\t" + "I-shiti1" + "\n"
                                tokenlist.append(eachlist[i+2+j])
                            else:
                                if j == 0:
                                    tokenlist[-1] = tokenlist[-1].replace("B-shiti1", "S-shiti1", 1)
                                else:
                                    tokenlist[-1] = tokenlist[-1].replace("I-shiti1", "E-shiti1", 1)
                                count = i+2+j+1
                                break
                else:
                    eachlist[i] = eachlist[i] + "\t" + "O" + "\n"
                    tokenlist.append(eachlist[i])
                    count = count + 1
        tokenlist.append("\n")
    for token in tokenlist:
        writetxt_a(token, path)

def writetxt_a(txt,path):
    with codecs.open(path,'a','utf-8') as w:
        w.write(txt)
    
        
path="O:\\大三下\\中文信息处理实验"
path1="O:\\大三下\\中文信息处理\\1.txt"
fileArray=fun(path)
for raw_file_path in fileArray:
    raw_file=readfile(raw_file_path)
    print(raw_file[1])
    for i in range(len(raw_file)):
        sentense = raw_file[i].split("。")
        writetoken2txt(sentense,path1)
```
