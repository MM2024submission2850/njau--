```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
%matplotlib inline  #设置图像显示方式
sns.set()    #设置画图空间为 Seaborn 默认风格

# plt.rcParams['axes.unicode_minus'] = False # 解决负号'-'显示为方块的问题
```

```python
df=pd.read_csv(r'C:\Users\13284\Desktop\data.csv',encoding = 'utf-8')
df
```

```python
#查看男女的占比
print(df['性别'].value_counts())

fig,ax = plt.subplots(1,2,figsize=(30,15))
sizes = [60,30,100]
# 用pandas画图
df['性别'].value_counts().plot.pie(ax=ax[0],shadow=False,autopct='%1.2f%%',textprops={'fontsize': 24})
ax[0].set_ylabel('',fontsize=20)    #设置y轴标签
ax[0].set_xlabel('性别',fontsize=20)    #设置x轴标签

# 用seaborn画图
sns.countplot('性别',data=df,ax=ax[1])
ax[1].set_ylabel('',fontsize=20)
ax[1].set_xlabel('性别',fontsize=20)
plt.show()
```

```python
#对从业时间进行预处理
job=df['从业时间'].str.split('-',expand=True)[0]
#对出生年份进行预处理
df['出生日期'] = df['出生日期'].astype(str)
born=df["出生日期"].str[0:4]
#转化int
job=job.astype(int)
born=born.astype(int)
print(job,born)

```
