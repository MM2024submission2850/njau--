```python
#回文数的问题
s=input().split()
a=int(s[0])
b=int(s[1])
n=0
for a in range (a,b+1):
    l=str(a)
    list_x=list(l)
    list_x.reverse()
    y="".join(list_x)
    if y==l:
        n=n+1
print(n)
```
