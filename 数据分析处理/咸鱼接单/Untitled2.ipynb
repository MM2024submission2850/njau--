```python
#查看区域的占比
print(df['区域'].value_counts())

fig,ax = plt.subplots(1,2,figsize=(30,15))
sizes = [60,30,100]
# 用pandas画图
df['区域'].value_counts().plot.pie(ax=ax[0],shadow=False,autopct='%1.2f%%',textprops={'fontsize': 24})
ax[0].set_ylabel('',fontsize=20)    #设置y轴标签
ax[0].set_xlabel('区域',fontsize=20)    #设置x轴标签

# 用seaborn画图
sns.countplot('区域',data=df,ax=ax[1])
ax[1].set_ylabel('',fontsize=20)
ax[1].set_xlabel('区域',fontsize=20)
plt.show()
```

```python
#再来看看不同区的房价

bostonCHAS = df[['区域','单价（平方米）']]    #先将CHAS和PRICE两列数据取出

bostonCHAS1=bostonCHAS.pivot_table(values='单价（平方米）',    #计算的值
                               index='区域',       #透视的行，分组的依据
                               aggfunc='mean')            #聚合函数

# 对透视表进行降序排列
bostonCHAS1 = bostonCHAS1.sort_values(by='单价（平方米）',     # 排序依据
                        ascending=False                 # 是否升序排列
                       )

bostonCHAS1.plot(kind='bar',title ="上海市行政区房价梯队",figsize=(15,10),legend=True, fontsize=12)
```

```python
#相关系数计算
corrboston = df.corr()
print(corrboston)

plt.figure(figsize=(10,10))    #设置画布
sns.heatmap(corrboston,annot=True,cmap='RdGy')
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus'] = False
plt.show()
```

```python
#查看区域的占比
print(df['户型'].value_counts())

fig,ax = plt.subplots(1,2,figsize=(30,15))
sizes = [60,30,100]
# 用pandas画图
df['户型'].value_counts().plot.pie(ax=ax[0],shadow=False,autopct='%1.2f%%',textprops={'fontsize': 24})
ax[0].set_ylabel('',fontsize=20)    #设置y轴标签
ax[0].set_xlabel('户型',fontsize=20)    #设置x轴标签

# 用seaborn画图
sns.countplot('户型',data=df,ax=ax[1])
ax[1].set_ylabel('',fontsize=20)
ax[1].set_xlabel('户型',fontsize=20)
plt.show()
```
